````python
# 加载API_KEY，请在.env中自行配置。
import os
from dotenv import load_dotenv
load_dotenv()
API_KEY = os.getenv("API_KEY")

# pip install openai
from openai import OpenAI
client = OpenAI(
    api_key=API_KEY,  # Access Token属于个人账户的重要隐私信息，请谨慎管理，切忌随意对外公开,
    base_url="https://aistudio.baidu.com/llm/lmapi/v3",  # aistudio 大模型 api 服务域名
)

# 调用模型，生成回答
def get_response(q, model="1.ernie-x1.1-preview"):
    # 简单评测
    if model == "1.ernie-x1.1-preview":
        chat_completion = client.chat.completions.create(
            model="ernie-x1.1-preview",
            messages=[
            {
                "role": "user",
                "content": q
            },
        ],
            stream=True,
            extra_body={
                "web_search": {
                    "enable": True
                }
            },
            max_completion_tokens=65536
        )
    

    # 模型对比
    if model == "2.ernie-x1.1-preview":
        chat_completion = client.chat.completions.create(
            model="ernie-x1.1-preview",
            messages=[
            {
                "role": "system",
                "content": "将有一些问题对，包含A句子和B句子，请逐一判断A句子、B句子是否是同一句话，并逐一打分(0-100)分。请无需过程，直接返回平均分。"
            },
            {
                "role": "user",
                "content": q
            },
        ],
            stream=True,
            extra_body={
                "web_search": {
                    "enable": True
                }
            },
            max_completion_tokens=65536
        )
    if model == "2.kimi-k2-instruct":
        chat_completion = client.chat.completions.create(
            model="kimi-k2-instruct",
            messages=[
            {
                "role": "system",
                "content": "将有一些问题对，包含A句子和B句子，请逐一判断A句子、B句子是否是同一句话，并逐一打分(0-100)分。请无需过程，直接返回平均分。"
            },
            {
                "role": "user",
                "content": q
            },
        ],
            stream=True,
            extra_body={

            },
            max_completion_tokens=32768,
            temperature=0.6,
            top_p=1
        )
    if model == "2.deepseek-v3":
        chat_completion = client.chat.completions.create(
            model="deepseek-v3",
            messages=[
            {
                "role": "system",
                "content": "将有一些问题对，包含A句子和B句子，请逐一判断A句子、B句子是否是同一句话，并逐一打分(0-100)分。请无需过程，直接返回平均分。"
            },
            {
                "role": "user",
                "content": q
            },
        ],
            stream=True,
            extra_body={
                "web_search": {
                    "enable": True
                }
            },
            max_completion_tokens=8000,
            temperature=0.6
        )
    if model == "2.qwen3-235b-a22b":
        chat_completion = client.chat.completions.create(
            model="qwen3-235b-a22b",
            messages=[
            {
                "role": "system",
                "content": "将有一些问题对，包含A句子和B句子，请逐一判断A句子、B句子是否是同一句话，并逐一打分(0-100)分。请无需过程，直接返回平均分。"
            },
            {
                "role": "user",
                "content": q
            },
        ],
            stream=True,
            extra_body={
                "penalty_score": 1,
                "enable_thinking": True
            },
            max_completion_tokens=8192,
            temperature=0.7,
            top_p=0.8
        )
    if model == "2.llama-4-maverick-17b-128e-instruct":
        chat_completion = client.chat.completions.create(
            model="llama-4-maverick-17b-128e-instruct",
            messages=[
            {
                "role": "system",
                "content": "将有一些问题对，包含A句子和B句子，请逐一判断A句子、B句子是否是同一句话，并逐一打分(0-100)分。请无需过程，直接返回平均分。"
            },
            {
                "role": "user",
                "content": q
            },
        ],
            stream=True,
            extra_body={
                "penalty_score": 1
            },
            max_completion_tokens=8192,
            temperature=0.6,
            top_p=0.9
        )


    # 结果输出
    print("-"*100)
    n = 0
    for chunk in chat_completion:
        if not chunk.choices or len(chunk.choices) == 0:
            continue
        
        if hasattr(chunk.choices[0].delta, "reasoning_content") and chunk.choices[0].delta.reasoning_content:
            # print(chunk.choices[0].delta.reasoning_content, end="", flush=True)
            pass
        else:
            if n == 0:
                n = 1
            if chunk.choices[0].delta.content:
                print(chunk.choices[0].delta.content, end="", flush=True)
    print("\n"+"-"*100)
````
